```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded["player_of_the_week (1).csv"]))
```

```python
df
```

```python
#title Conference vs Seasons in league

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(df['Conference'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df, x='Seasons in league', y='Conference', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# What is the most and the least (still in activity) team rewarded per week in the NBA from 1979 to 2018?


# Count of Players rewarded POTW per Team

df['Counts'] = df.groupby(['Team'])['Player'].transform('count')
plt.bar(df["Team"], df["Counts"], color = ["darkkhaki"])
plt.xticks(rotation=90)
plt.show()
```

The Los Angeles Lakers is the team with most players awarded, from 1979 to 2018, with 88. While the San Diego Clippers is the least team with only 1 playes.

```python

# What is the most and the least position rewarded per week in the NBA from 1979 to 2018?
# Count of Players rewarded POTW per Position

df['Counts'] = df.groupby(['Position'])['Player'].transform('count')
plt.bar(df["Position"], df["Counts"], color = ["slategrey"])
plt.xticks(rotation=90)
plt.show()
```

The Position G (Guard) is the most rewarded position in the NBA with 200 Players in totatl while the G-F (Guard-Forward) is the least rewarded with 3 Players
